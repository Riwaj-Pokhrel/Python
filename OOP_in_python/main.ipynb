### Access modifiers: protected attributes

```python
class User:
    def __init__(self,username,email,password):
        self.username=username
        self._email=email #doen't actually mean protected. It is just a sign for developers to say that it should be protected.
        self.password=password
    def get_email(self):
        return self._email
user1=User('Ram','ram@gmail.com','123')
print(user1._email) #gives the output as python doesn't enforce strict rules regarding access of attributes
user1._email='shyamkumari'
print(user1._email)
print(user1.get_email()) #it should be the right way to do it.
```

### Getters and Setters

```python
from datetime import datetime
class Person:
    def __init__(self,username,email,password):
        self.username=username
        self._email=email
        self.password=password
    def get_email(self):
        print('Email accessed at:',datetime.now())
        return self._email
    def set_email(self,new_email):
        if '@'in new_email:
            self._email=new_email
        else:
            print("Invalid new_email.")
user1=Person('Ram','ram@gmail.com','123')
print(user1.get_email())
user1.set_email('shyamgmail.com')
print(user1.get_email())
```

### Properties for getting and setting attributes

```python
class User:
    def __init__(self,username,email,password):
        self.username=username
        self._email=email
        self.password=password
    @property
    def email(self):
        print('Email accessed')
        return self._email
    @email.setter
    def email(self,new_email):
        if"@"in new_email:
            self._email=new_email
        else:
            print("Invalid Email.")

user1=User('Ram','ram@gmail.com','123')
print(user1.email)
user1.email='shyamkumar'
print(user1.email)
```

### Static attributes

```python
class User:
    user_count=0 
    def __init__(self,username,email):
        self.username=username
        self.email=email
        User.user_count+=1
    def display_user(self):
        print(f"Username:{self.username},Email:{self.email}")
    def __del__(self):
        User.user_count-=1

user1=User('Ram','ram@gmail.com')
print(User.user_count)
user2=User('Shyam','shyamkumari@gmail.cm')
print(user1.user_count)
del user2
print(User.user_count)
```

### Static Methods

```python
class BankAccount:
    MIN_BALANCE = 100

    def __init__(self, owner, balance=0):
        self.owner = owner
        self._balance = balance

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"{self.owner}'s new balance: ${self._balance}")
        else:
            print("Deposit amount must be positive.")

    @staticmethod
    def is_valid_interest_rate(rate):
        return 0 <= rate <= 5

    
account1 = BankAccount('Ram', 600)
account1.deposit(200)
print(BankAccount.is_valid_interest_rate(4.5))
```
