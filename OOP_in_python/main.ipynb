### Getters and Setters

```python
from datetime import datetime
class Person:
    def __init__(self,username,email,password):
        self.username=username
        self._email=email
        self.password=password
    def get_email(self):
        print('Email accessed at:',datetime.now())
        return self._email
    def set_email(self,new_email):
        if '@'in new_email:
            self._email=new_email
        else:
            print("Invalid new_email.")
user1=Person('Ram','ram@gmail.com','123')
print(user1.get_email())
user1.set_email('shyamgmail.com')
print(user1.get_email())
```

### Properties for getting and setting attributes

```python
class User:
    def __init__(self,username,email,password):
        self.username=username
        self._email=email
        self.password=password
    @property
    def email(self):
        print('Email accessed')
        return self._email
    @email.setter
    def email(self,new_email):
        if"@"in new_email:
            self._email=new_email
        else:
            print("Invalid Email.")

user1=User('Ram','ram@gmail.com','123')
print(user1.email)
user1.email='shyamkumar'
print(user1.email)
```

### Static Methods

```python
class BankAccount:
    MIN_BALANCE = 100

    def __init__(self, owner, balance=0):
        self.owner = owner
        self._balance = balance

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"{self.owner}'s new balance: ${self._balance}")
        else:
            print("Deposit amount must be positive.")

    @staticmethod
    def is_valid_interest_rate(rate):
        return 0 <= rate <= 5

    
account1 = BankAccount('Ram', 600)
account1.deposit(200)
print(BankAccount.is_valid_interest_rate(4.5))
```

```python
class BankAccount:
    MIN_BALANCE = 100

    def __init__(self, owner, balance=0):
        self.owner = owner
        self._balance = balance

    def deposit(self, amount):
        if self._is_valid_amount(amount):
            self._balance += amount
            print(f"{self.owner}'s new balance: ${self._balance}")
        else:
            print("Deposit amount must be positive.")

    @staticmethod
    def is_valid_interest_rate(rate):
        return 0 <= rate <= 5
    def _is_valid_amount(self,amount): #protected method #similaryl __func for private method.
        return amount>0
    
account1 = BankAccount('Ram', 600)
account1.deposit(200)
print(BankAccount.is_valid_interest_rate(4.5))
```

### Encapsulation

```python
class BankAccount:
    def __init__(self,balance):
        self._balance=balance
    @property
    def balance(self):
        return self._balance
    def deposit(self,amount):
        if amount<=0:
            raise ValueError('Deposit amount must be positive.')
        self._balance+=amount
    def withdraw(self,amount):
        if amount<=0:
          raise ValueError('Withdraw amount must be positive.')
        if amount> self.balance:
            raise ValueError("Insufficient funds.")
        self._balance-=amount

account=BankAccount(1000)
print(account.balance)
account.deposit(199)
print(account.balance)
account.withdraw(99)
print(account.balance)
```

### Abstraction
Reduce complexity by hiding unnecesssary details

```python
class EmailService:
    def _connect(self):
        print('Connecting to email server')
    
    def _authenticate(self):
        print('Authenticating....')
    
    def send_email(self):
        self._connect()
        self._authenticate()
        print('Sending emial...')
        self._disconnect()
    def _disconnect(self):
        print('Disconnecting from email server ...')

email=EmailService()
email.send_email()
```

### Inheritance
creating of new classes (sub classes or derived classes ) based on existing classes (superclasses or base classes).

```python
class Vehicle:
    def __init__(self,brand,model,year):
        self.brand=brand
        self.model=model
        self.year=year
    def start(self):
        print('Vehicle is starting.')
    def stop(self):
        print('Vehicle is stopping.')

class Car(Vehicle):
    def __init__(self, brand, model, year,no_of_wheels):
        super().__init__(brand, model, year)
        self.no_of_wheels=no_of_wheels
class Bike(Vehicle):
    def __init__(self, brand, model, year,no_of_wheels):
        super().__init__(brand, model, year)
        self.no_of_wheels=no_of_wheels
car=Car("Ford","Focus",2008,4)
bike=Bike('Honda','Scoopy','2018',2)
print(car.__dict__)
print(bike.__dict__)
car.start()
```

### Polymorphism

```python
# Naive Solution
class Car:
    def __init__(self,brand,model,year,no_of_wheels):
        self.brand=brand
        self.model=model
        self.year=year
        self.no_of_wheels=no_of_wheels
    def start_car(self):
        print('Car is starting.')
    def stop_car(self):
        print('Car is stopping.')


class Motorcycle:
    def __init__(self,brand,model,year,no_of_wheels):
        self.brand=brand
        self.model=model
        self.year=year
        self.no_of_wheels=no_of_wheels
    def start_bike(self):
        print('Motorcycle is starting.')
    def stop_bike(self):
        print('Motorcycle is stopping.')

vehicles=[
    Car("Ford","Focus",2008,4),
    Motorcycle('Honda','Scoopy',2018,2)
]

for vehicle in vehicles:
    if isinstance(vehicle,Car):
        print(f"Inspecting {vehicle.brand} {vehicle.model}({type(vehicle).__name__})")
        vehicle.start_car()
        vehicle.stop_car()
    elif isinstance(vehicle,Motorcycle):
        print(f'Inspecting {vehicle.brand} {vehicle.model}({type(vehicle).__name__})')
        vehicle.start_bike()
        vehicle.stop_bike()
    else:
        raise Exception("Object is not found in vehicles.")
```
