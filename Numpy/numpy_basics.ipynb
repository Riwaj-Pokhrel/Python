### Creating numpy arrays

```python
import numpy as np
a=np.array([1,2,3,4,5])
print(a)
print(type(a))
```

```python
# 2D and 3D 
b=np.array([[1,2,3],[4,5,6]])
print(b)
print(type(b))

c=np.array([[[1,2],[3,4]],[[4,5],[6,7]]])
print(c)
```

```python
# dtype
np.array([1,2,3],dtype=complex)
np.array([1,2,3],dtype=float)
```

```python
# np.arange
np.arange(1,11)
```

```python
np.arange(1,20,2).reshape(5,2)
```

```python
# np.ones and np.zeros
np.ones((3,4))
```

```python
np.zeros((4,6))
```

```python
np.random.random((4,2))
```

```python
# np.linspace
np.linspace(-10,10,10)
```

```python
np.identity(3)
```

### Array Operations

```python
a1=np.arange(12).reshape(3,4)
a2=np.arange(12,24).reshape(3,4)
```

```python
# scalar operations:
a1 *2
```

```python
a1>5
```

```python
# Vector operations:
a1+a2
```

### Array Functions

```python
a1=np.random.random((3,3))
a1=np.round(a1*100)
a1
```

```python
# min / max
np.max(a1)
np.min(a1)
```

```python
#sum / prod
np.sum(a1)
np.prod(a2)
```

```python
#0-> column and 1->row
np.max(a1,axis=1)
```

```python
np.sum(a1,axis=1)
```

```python
# mean/median/std/var
print(np.mean(a1))
print(np.median(a1))
print(np.std(a1))
print(np.var(a1))
```

```python
# trigonometric functions
np.sin(a1)
```

```python
# dot product
a2=np.arange(12).reshape(3,4)
a3=np.arange(12,24).reshape(4,3)
np.dot(a2,a3)
```

```python
# log and exponents 
np.log(a1)
```

```python
np.exp(a1)
```
