### Arithmetic Operators in Python

```python
print('Addition: ', 1 + 2)        
print('Subtraction: ', 2 - 1)     
print('Multiplication: ', 2 * 3)  
print ('Division: ', 4 / 2)       # Division in Python gives floating number
print('Division: ', 6 / 2)        # 3.0         
print('Division: ', 7 / 2)        # 3.5
print('Floor Division:', 7 // 2)  #floor division(gives integer)
print ('Floor Division:',7 // 3)   # 2
print('Modulus: ', 3 % 2)         
print('Exponentiation: ', 2 ** 3) 
print('Anding:',3&4)
print('Oring:',3|4)
print('Xoring:',3^4)
print('Left shift:',5<<2) #shifts bits of 5 left by 2 positions , equivalent to 5*(2^2)
print('Right shift:',26>>3) #equivalent to 26//8(2^3)
print('Addition of complex numbers:',(5+2j)+(4+23j))
```

### Comparison Operators

```python
print(3 > 2)  
print(3 >= 2)
print(3 < 2) 
print(2 < 3) 
print(2 <= 3)
print(3 == 2)
print(3 != 2)
print(len('mango') == len('avocado')) 
print(len('mango') != len('avocado'))  
print(len('mango') < len('avocado'))   
print('True == True: ', True == True)
print('True > False: ', True > False)
print('False == False:', False == False)
```

```python
print('1 is 1', 1 is 1)                   
print('1 is not 2', 1 is not 2) 
print('Riwaj is Riwaj','Riwaj'is'Riwaj') 
print('Riwaj is Riwaj_Pokhrel','Riwaj'is'Riwaj_Pokhrel')           
print('R in Riwaj', 'R' in 'Riwaj') 
print('P in Riwaj', 'P'in 'Riwaj') 
print('coding' in 'coding for all') 
print('a in an:', 'a' in 'an')     
print('4 is 2 ** 2:', 4 is 2 ** 2) 
```

```python
print(3 > 2 and 4 > 3) 
print(3 > 2 and 4 < 3) 
print(3 < 2 and 4 < 3) 
print('True and True: ', True and True)
print(3 > 2 or 4 > 3)  
print(3 > 2 or 4 < 3)  
print(3 < 2 or 4 < 3)  
print('True or False:', True or False)
print(not 3 > 2)
print(not True) 
print(not False)
print(not not True)  
print(not not False) 
```
