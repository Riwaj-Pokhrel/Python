### Creating a Tuple


```python
empty_tuple=()
print(empty_tuple)
print(len(empty_tuple))

empty_tuple=tuple()
print(empty_tuple)
print(len(empty_tuple))
```

```python

fruits = ('banana', 'orange', 'mango', 'lemon')
print(fruits)
print(len(fruits))
first_fruit = fruits[0]
second_fruit = fruits[1]
last_index =len(fruits) - 1
last_fruit = fruits[last_index]
print(first_fruit)
print(second_fruit)
print(last_fruit)
print('Last fruit is ',fruits[-1])
```

### Slicing tuples


```python
fruits = ('banana', 'orange', 'mango', 'lemon')
print(fruits[0:3])
print(fruits[-4:])
print(fruits[::-1])
```

```python
fruits = ('banana', 'orange', 'mango', 'lemon')
fruits_list=list(fruits)
fruits_list.reverse()
print(fruits_list)
fruits_tuple=tuple(fruits_list)
print(fruits_tuple)
```

```python
fruits = ('banana', 'orange', 'mango', 'lemon')
print('banana'in fruits)
print(fruits.count('banana'))
print(fruits.index('banana'))
```

```python
fruits = ('banana', 'orange', 'mango', 'lemon')
vegetables = ('Tomato', 'Potato', 'Cabbage','Onion', 'Carrot')
fruits_and_vegetables = fruits + vegetables
print(len(fruits))
print(len(vegetables))
print(fruits_and_vegetables)
print(len(fruits_and_vegetables))
fruits,vegetables=fruits_and_vegetables[0:4],fruits_and_vegetables[4:]
print(fruits)
print(vegetables)
```
