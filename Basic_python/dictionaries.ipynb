### Creating Dictionaries

```python
empty_dict={}
print(empty_dict)
print(type(empty_dict))
print(len(empty_dict))

empty_dict=dict()
print(empty_dict)
print(type(empty_dict))
print(len(empty_dict))
```

```python
info={'first_name':'Riwaj','last_name':'Pokhrel','faculty':'BCT'}
print(info)
#the key or value can be any data type
numbers={1:'Ram',2:3,4:5.67,6:('Ram','Shyam',6,8.90,[12,34,56,89])}
print(numbers)
print(len(numbers))
```

### Accesing Dictionary items

```python
info={'first_name':'Riwaj','last_name':'Pokhrel','faculty':'BCT'}
print('First Name:',info['first_name'])#gives error if the key doesn't exist
print('Age:',info.get('age'))
```

### Adding items to Dictionary

```python
info={'first_name':'Riwaj','last_name':'Pokhrel','faculty':'BCT','skills':['html','css','js','python']}
info['age']=20
print(info)
info['skills'].append('react')
print(info)
```

### Modifying Dictionary items

```python
info={'first_name':'Riwaj','last_name':'Pokhrel','faculty':'BCT','skills':['html','css','js','python']}
info['faculty']='bct'
print(info)
info['skills[0]']='HTML'#doesn't work this way
print(info)
info['skills'][0]='HTML'#correct way
print(info)
```

```python
info={'first_name':'Riwaj','last_name':'Pokhrel','faculty':'BCT','skills':['html','css','js','python']}
print('skills'in info)
```

### Removing Key and value pairs from a dictionary

```python
info={'first_name':'Riwaj','last_name':'Pokhrel','faculty':'BCT','skills':['html','css','js','python']}
info.pop('faculty')
info.popitem()
print(info)
del info['last_name']
print(info)
info_copy=info.copy()
del info
# print(info) will give error
print(info_copy)
info_copy.clear()
print(info_copy)
```

### Changing Dictionary to a list of items

```python
info={'first_name':'Riwaj','last_name':'Pokhrel','faculty':'BCT','skills':['html','css','js','python'],'age':20}
print(info.items())
print(info.keys())
print(info.values())
print(type(info['skills']))
print(type(info['age']))
```
