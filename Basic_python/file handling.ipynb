### Writing to the file

```python
#Case 1-if the file is not present.
f=open('/files/file1.txt','w')
f.write('This is a file handling example')
f.close()
```

```python
# Writing multiple lines in a file
f=open('/files/Sample1.txt','w')
f.write('Hello World!')
f.write('\n How are you?')
f.close()
 
```

```python
#case 2 -if the file is already present
f=open('/files/file1.txt','w')
f.write('Writing in a already existing file')
f.close()
```

```python
# introducing append mode
f=open('/files/Sample1.txt','a')
f.write('\nAppending to the end of the file.')
f.close()
```

```python
# writing lines .
L=['\nHello,','How are you?\n','-Hello.I am fine']
f=open('/files/file1.txt','a')
f.writelines(L)
f.close()
```

### Reading  files

```python
# reading from files
f=open('/files/file1.txt','r')
s=f.read()
f.close()
print(s)
```

```python
#reading upto n characters
f=open('/files/file1.txt','r')
s=f.read(10)
f.close()
print(s)
```

```python
#read line -to read line by line
f=open(' /files/file1.txt','r')
print(f.readline())
print(f.readline())
f.close()
```

### Using Context Manager(With)

```python
with open(' /files/file2.txt','w') as f:
    f.write('Writing to the file using with. With helps to close the file as the usage is over.')
```

```python
with open(' /files/file2.txt','r') as f:
    print(f.read())
```

```python
with open(' /files/file2.txt','r') as f:
    print(f.read(10))
    print(f.read(10))
```

### Seek and tell functions

```python
with open(' /files/file2.txt','r') as f:
    print(f.read(24))
    print(f.tell())
    f.seek(34)
    print(f.read(10))
```

### Serialization and deserialization of class objects

```python
class Person:
    def __init__(self,fname,lname,age,gender):
        self.fname=fname
        self.lname=lname
        self.age=age
        self.gender=gender

person=Person('Ram','Kumar',33,'Male')

    
```

```python
# As a string
def show_object(person):
    if isinstance(person,Person):
        return "{}{} age->{} gender->{}".format(person.fname,person.lname,person.age,person.gender)
    
with open('details.json','w')as f:
    json.dump(person,f,default=show_object)
```

```python
with open(' /files/details.json','r')as f:
    print(json.load(f))
```

### Pickling
- The pickle module in Python is used for serializing and de-serializing Python objects. This process, also known as "pickling" and "unpickling", converts a Python object hierarchy into a byte stream, which can then be stored in a file, a database, or sent over a network, and later reconstructed back into the original Python object. 

```python
class Person:
    def __init__(self,fname,lname,age):
        self.fname=fname
        self.lname=lname
        self.age=age
    def display_info(self):
        print(f'Hi! I am {self.fname} {self.lname}.I am {self.age} years old.')

person=Person('Ram','Kumar',33) 
```

```python
import pickle
```

```python
with open('details2.pkl','wb')as f:
    pickle.dump(person,f)
```

```python
with open('details2.pkl','rb')as f:
    p=pickle.load(f)
p.display_info()
```
