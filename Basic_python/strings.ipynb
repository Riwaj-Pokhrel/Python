### Strings in Python

```python
letter = 'P'               
print(letter)              
print(len(letter))         
greeting = 'Hello, World!' 
print(greeting)            
print(len(greeting))       
sentence = "I am learning python"
print(sentence)
print(len(sentence))
multiline_string = '''I am currently learning Python.
I enjoy learning new things.'''
print(multiline_string)
print (len(multiline_string))
```

### String Concatenation

```python
first_name='Riwaj'
last_name='Pokhrel'
full_name=first_name+last_name
print(full_name)
full_name=first_name,last_name
print(full_name)
full_name=first_name+' '+last_name
print(full_name)

```

### Escape Sequences in strings

### Formatting Strings

```python
first_name = 'Riwaj'
last_name = 'Pokhrel'
language = 'Python'
formated_string="I am %s %s.I am learining %s."%(first_name,last_name,language)
print(formated_string)
formated_string="I am {} {}.I am learining {}".format(first_name,last_name,language)
print(formated_string)
```

```python
radius = 10
pi = 3.14
area = pi * radius ** 2
formated_string = 'The area of circle with a radius %d is %.2f.' %(radius, area)
print(formated_string)
formated_string = 'The area of a circle with a radius {} is {:.2f}.'.format(radius, area)
print(formated_string)
```

```python
python_libraries = ['Django', 'Flask', 'NumPy', 'Matplotlib','Pandas']
formated_string = 'The following are python libraries:%s' % (python_libraries)
print(formated_string) 
```

```python
#String interpolation
a = 4
b = 3
print(f'{a} + {b} = {a +b}')
print(f'{a} - {b} = {a - b}')
print(f'{a} * {b} = {a * b}')
print(f'{a} / {b} = {a / b:.2f}')
print(f'{a} % {b} = {a % b}')
print(f'{a} // {b} = {a // b}')
print(f'{a} ** {b} = {a ** b}')
print(f'My name is {first_name} {last_name}')
```

```python
language = 'Python'
a,b,c,d,e,f = language # unpacking sequence characters into variables
print(a) # P
print(b) # y
print(c) # t
print(d) # h
print(e) # o
print(f) # n
```

```python
language = 'Python'
first_letter = language[0]
print(first_letter) # P
second_letter = language[1]
print(second_letter) # y
last_index = len(language) - 1
last_letter = language[last_index]
print(last_letter) # n
```

### String Methods

```python
activity='i am learning python'
print(activity.capitalize())
print(activity.count('a'))
print(activity.count('learn'))
print(activity.count(' '))
print(activity.endswith('on'))

```

```python
activity2='I\tam\tlearning\tpython'
print(activity2)
activity2.expandtabs(19)
print(activity2)
print(activity2.expandtabs(14))

#find(): Returns the index of the first occurrence of a substring, if not found returns -1
#rfind(): Returns the index of the last occurrence of a substring, if not found returns -1
print(activity.find('yth'))
print(activity.rfind('i'))



```

```python
print(activity.index('am'))
print(activity.index('i',3))
print(activity.rindex('i'))
```

```python
activity='IamleariningPython'
print(activity.isalnum())       
activity='IamleariningPython99'
print(activity.isalnum()) 
activity='I am learining Python 99'#space is not alphanumeric
print(activity.isalnum()) 

#similarly isalpha() , isdecimal()
```

```python
pushups= 'Thirty'
print(pushups.isdigit()) 
pushups= '30'
print(pushups.isdigit()) 
pushups= '\u00B2'          #isdecimal doesn't accept unicode for numbers but it does
print(pushups.isdigit()) 
#isnumeric is simailar to is digit
#we also have islower() , isupper()
```
