### Variables in Python

```python
first_name='Riwaj'
last_name='Pokhrel'
country='Nepal'
age=20
skills=['HTML','CSS','C','C++','Python']
personal_info={
'first name':'Riwaj',
'last name':'Pokhrel',
'age':20,
'country':'Nepal'
}
```

### Built in functions


```python
print('Hello World')
print('Hello','World'+'!')
print(len('Hello World'))
Name=input('Enter Your Name') #always takes the input as string
print(Name)
print(type(Name))
Age=int(input('Enter Your age'))
print(Age)
print(type(age))
```

```python
print('First name:', first_name)
print('First name length:', len(first_name))
print('Last name:', last_name)
print('Last name length:',len(last_name))
print('Country:', country)
print('Age:', age)
print('Skills:', skills)
print('Person information: ', personal_info)
```

### Declaring Multiple variables in a line

```python
first_name,last_name,age='Riwaj','Pokhrel',20
print('First Name:',first_name)
print('Last name: '+last_name)
print('Age:',age)
print(len(first_name))
```

Type Casting


```python
#int to float
num_int=10
print('num_int:',num_int)
num_float=float(num_int)
print('num_float:',num_float)
```

```python
# float to int
gravity = 9.81
print(int(gravity))  
```

```python
# int to str
num_int = 10
print(str(num_int))
print(type(num_int))
str=str(num_int)
print(type(str)) 
```

```python
#str to int or float
num_str='109.87'
# num_int=int(num_str)
# print(num_int)            gives error
print(type(num_str))
num_float=float(num_str)
print(num_float)
print(type(num_float))
num_int=int(num_float)
print(num_int)
print(type(num_int))
```

```python
#str to list
str='Riwaj'
str_to_list=list(str)
print(str_to_list)
print(type(str_to_list))
```
