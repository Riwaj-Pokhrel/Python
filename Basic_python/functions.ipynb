### Function returning a value

```python
def full_name():
    first_name='Riwaj'
    Last_name='Pokhrel'
    full_name=first_name+' '+Last_name
    return full_name
print('Name:',full_name())
```

### Functions with parameters


```python
def activity(name,language):
    message='My name is %s. I am learining %s.'%(name,language)
    return message
print('Message:',activity('Riwaj','Python'))
```

```python
def weight_of_object (mass, gravity):
    weight = mass*gravity
    return weight
print('Weight of an object in Newtons: ',str(weight_of_object(100, 9.81))+'N')
```

### Passing Arguements with key and value

```python
def print_fullname(firstname, lastname):
    space = ' '
    full_name = firstname  + space + lastname
    return full_name
print(print_fullname(lastname='Pokhrel',firstname='Riwaj'))
```

```python
def returning_none():
    print('If we don\'t return anything from a function, it will return none automatically.')
print(returning_none())

```

```python
def is_even (n):
    if n % 2 == 0:
        print('even')
        return True    # return stops further execution of the function, similar to break 
    return False
print(is_even(10)) 
print(is_even(7)) 
```

```python
def find_evens(n):
    evens=[]
    for i in range(n+1):
        if i%2==0:
            evens.append(i)
    return evens
print('The even numbers are:',find_evens(20))
        
```

### Arbitary number of arguements
If we do not know the number of arguments we pass to our function, we can create a function which can take arbitrary number of arguments by adding * before the parameter name.

```python
def sum_all_num(*nums):
    total=0
    for i in nums:
        total+=i
    return total
print('The sum is:',sum_all_num(23,45,67))
```

### Default and Arbitrary Number of Parameters in Functions

```python
def numbers(first_num,*all):
    print(first_num)
    for i in all:
        print(i)
numbers(1,2,3,4,5,6,7,8)
```

### Function as a parameter of another function


```python
def square_of_sum(total):
    return total**2
def add(num1,num2):
    return num1+num2
print('The square of sum is:',square_of_sum(add(2,2)))
```
