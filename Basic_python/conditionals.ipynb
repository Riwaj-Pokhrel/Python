```python
month=input('Enter a month')
month=month.lower()
autumn=['september','october','nobemver']
winter=['december','january','february']
spring=['march','april','may']
summer=['june','july','august']
if month in autumn:
    print('Its autumn')
elif month in winter:
    print('Its winter')
elif month in spring:
    print('Its spring')
elif month in summer:
    print('Its summer.')
else:
    print('Invalid month')
```
